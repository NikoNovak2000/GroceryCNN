```python
# LOADING DATA
import os
import requests
from zipfile import ZipFile

# URL to the zip file of the dataset
url = "https://github.com/marcusklasson/GroceryStoreDataset/archive/refs/heads/master.zip"

# Download the file
response = requests.get(url)
with open("GroceryStoreDataset.zip", "wb") as file:
    file.write(response.content)

# Extract the zip file
with ZipFile("GroceryStoreDataset.zip", "r") as zip_ref:
    zip_ref.extractall("GroceryStoreDataset")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the path to the dataset
dataset_path = 'GroceryStoreDataset/GroceryStoreDataset-master/dataset'

# Create an ImageDataGenerator instance for training and validation
datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values to [0,1]
    validation_split=0.2  # 20% of data will be used for validation
)

# Load the training data
train_generator = datagen.flow_from_directory(
    directory=dataset_path,
    subset='training',
    batch_size=32,
    target_size=(256, 256),  # Resize images to (256, 256)
    class_mode='sparse'  # Use sparse labels for classification
)

# Load the validation data
val_generator = datagen.flow_from_directory(
    directory=dataset_path,
    subset='validation',
    batch_size=32,
    target_size=(256, 256),
    class_mode='sparse'
)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
```

```python
data = tf.keras.utils.image_dataset_from_directory('dataset')
```

```python
data_iterator = data.as_numpy_iterator()
batch = data_iterator.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

```python
# SCALING DATA
data = data.map(lambda x,y: (x/255, y))
data.as_numpy_iterator().next()
```

```python
batch[1]
```

```python
# SPLIT DATA
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
# Deep Learning Model
train
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile('adam', loss=tf.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
# TRAIN
logdir='logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
# PlOT PERFORMANCE
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
# EVALUATE
import tensorflow as tf
from tensorflow.keras.metrics import Precision, Recall

# Initialize the metrics
pre = Precision()
re = Recall()

# Iterate over the test dataset
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    yhat_classes = tf.argmax(yhat, axis=-1)  # Convert probabilities to class labels
    
    # Update metrics
    pre.update_state(y, yhat_classes)
    re.update_state(y, yhat_classes)

# Print the results
print("Precision:", pre.result().numpy())
print("Recall:", re.result().numpy())
```

```python
test_loss, test_accuracy = model.evaluate(test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_predictions(generator, model):
    # Get a batch of images and labels
    images, labels = next(generator)
    
    # Make predictions
    predictions = model.predict(images)
    predicted_labels = np.argmax(predictions, axis=1)
    
    # If labels are already class indices, no need to apply argmax
    true_labels = labels.astype(int).flatten()  # Ensure labels are integers
    
    # Get class indices from the generator
    class_indices = {v: k for k, v in generator.class_indices.items()}
    
    # Plot images with true and predicted labels
    plt.figure(figsize=(10, 10))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].astype("uint8"))  # No need to call .numpy() on images
        true_label = class_indices.get(true_labels[i], "Unknown")
        pred_label = class_indices.get(predicted_labels[i], "Unknown")
        plt.title(f"True: {true_label}, Pred: {pred_label}")
        plt.axis("off")
    plt.show()

# Plot predictions using the validation generator
plot_predictions(val_generator, model)
```
